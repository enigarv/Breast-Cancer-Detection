```python
import fitz
import os
```

```python
path_pdfs = "/content/drive/MyDrive/Progetto Data Lab/Dati/PDF/"
path_images = "/content/drive/MyDrive/Progetto Data Lab/Dati/Immagini/"
```

```python
# Estrazione e salvataggio delle immagini contenute nei PDF tramite la libreria PyMuPDF
for file in sorted(os.listdir(path_pdfs)):
  print(file)
  if file.endswith('.pdf'):
    try:
      doc = fitz.open(path_pdfs + file)
      folder = file.split(".")[0]
      os.makedirs(path_images + folder)
      for i in range(len(doc)):
        for img in doc.get_page_images(i):
          xref = img[0]
          pix = fitz.Pixmap(doc, xref)
          if pix.n < 5:      
            pix.save(path_images + folder + "/p%s-%s.png" % (i, xref))
          else:               
            pix1 = fitz.save(fitz.csRGB, pix)
            pix1.save(path_images + folder + "/p%s-%s.png" % (i, xref))
            pix1 = None
          pix = None
    except:
      pass
```
