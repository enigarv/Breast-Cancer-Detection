```python
from PIL import Image
from collections import defaultdict
import matplotlib.pyplot as plt
from __future__ import print_function
import binascii
import struct
from PIL import Image
import numpy as np
import scipy
import scipy.misc
import scipy.cluster
import os
```

```python
NUM_CLUSTERS = 5
path_src = "/content/drive/MyDrive/Progetto Data Lab/Dati/Immagini/"
path_dst = "/content/drive/MyDrive/Progetto Data Lab/Dati/Immagini Viola/"
```

```python
# Selezione e salvataggio di tutte le immagini estratte dai PDF caratterizzate dalla predominanza del colore viola
for folder in sorted(os.listdir(path_src))[3000:]:
  for file in sorted(os.listdir(path_src + folder + "/")):
    img = Image.open(path_src + folder + "/" + file)
    im = img.resize((150, 150))
    ar = np.asarray(im)
    shape = ar.shape
    try:
      ar = ar.reshape(np.product(shape[:2]), shape[2]).astype(float)
      codes, dist = scipy.cluster.vq.kmeans(ar, NUM_CLUSTERS)
      vecs, dist = scipy.cluster.vq.vq(ar, codes)         
      counts, bins = np.histogram(vecs, len(codes))   
      index_max = np.argmax(counts)                    
      peak = codes[index_max]
      if ((peak[0] >= 125 and peak[1] <= 120 and peak[2] >= 125) or (peak[0] >= 150 and peak[1] <= 150 and peak[2] >= 150)):
        exists = os.path.exists(path_dst + folder)
        if not exists:
          os.makedirs(path_dst + folder)
        img.save(path_dst + folder + "/" + file)
    except:
      pass
```
