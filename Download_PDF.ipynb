```python
# Ricerca nel database PubMed di NCBI di tutte le pubblicazioni associate alla chiave di ricerca "lymph nodes metastasis breast cancer"
handle_all_articles = Entrez.esearch(db='pubmed', term='lymph nodes metastasis breast cancer', retmax = 100000) 
record_all_articles = Entrez.read(handle_all_articles)
print(record_all_articles)
```

```python
id_list = record_all_articles['IdList']
```

```python
# Estrazione e salvataggio dei PMID di tutte le pubblicazioni
pmid_list = []
for id in id_list:
  handle = Entrez.efetch(db='pubmed', id=id, retmode='xml')
  record = Entrez.read(handle)
  try:
    pmid = str(record['PubmedArticle'][0]['MedlineCitation']['PMID'])
    pmid_list.append(pmid)
  except:
    pass
```

```python
len(pmid_list)
```

```python
# Creazione di un file txt contenente tutti i PMID
textfile = open("pmid_list.txt", "w")
for pmid in pmid_list:
    textfile.write(pmid + "\n")
textfile.close()
```
